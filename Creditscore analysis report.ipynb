```python
indep=dataset[["Annual_Income","Num_Bank_Accounts","Monthly_Inhand_Salary","Interest_Rate","Num_of_Loan","Num_of_Delayed_Payment","Outstanding_Debt"]]
               
dep=dataset["Credit_Score"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size=1/3, random_state=42)
from sklearn.ensemble import RandomForestClassifier
grid = RandomForestClassifier()
grid.fit(X_train,y_train)
```

```python
#predict the xtest 
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
print(clf_report)
```
